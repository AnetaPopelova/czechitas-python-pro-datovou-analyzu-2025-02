## Kdy `[]`, `()`, a kdy nic?

| Syntaxe | Pou≈æit√≠ | P≈ô√≠klad |
|---------|----------------|----------------|
| `seznam[0]` | P≈ô√≠stup k prvku v seznamu | `my_list = [10, 20, 30]` ‚Üí `my_list[1]` ‚Üí `20` |
| `slovnik["klic"]` | P≈ô√≠stup k hodnotƒõ ve slovn√≠ku | `my_dict = {"jmeno": "Aneta"}` ‚Üí `my_dict["jmeno"]` ‚Üí `"Aneta"` |
| `text[3]` | P≈ô√≠stup k p√≠smenu v ≈ôetƒõzci | `"Python"[0]` ‚Üí `"P"` |
| `df["sloupec"]` | V√Ωbƒõr sloupce v pandas | `df["cena"]` |
| `df[["A", "B"]]` | V√Ωbƒõr v√≠ce sloupc≈Ø | `df[["cena", "produkt"]]` |
| `df[df["cena"] > 100]` | Filtrov√°n√≠ ≈ô√°dk≈Ø v pandas | Vybere ≈ô√°dky, kde je cena > 100 |
| `df.loc[5, "cena"]` | P≈ô√≠stup podle indexu a n√°zvu sloupce | Hodnota ve sloupci `cena` pro index 5 |
| `df.iloc[5, 2]` | P≈ô√≠stup podle pozice | Hodnota v 5. ≈ô√°dku a 2. sloupci |
| `print("Ahoj")` | Vol√°n√≠ funkce v Pythonu | `len("Python")` ‚Üí `6` |
| `my_list.append(4)` | Pou≈æit√≠ metody seznamu | P≈ôid√° `4` do seznamu |
| `df.head()` | Metoda v pandas | Zobraz√≠ prvn√≠ch 5 ≈ô√°dk≈Ø |
| `df.describe()` | Statistick√° sumarizace | Vr√°t√≠ popisnou statistiku tabulky |
| `df.shape` | Poƒçet ≈ô√°dk≈Ø a sloupc≈Ø | `(100, 5)` pro 100 ≈ô√°dk≈Ø, 5 sloupc≈Ø |
| `df.columns` | Seznam n√°zv≈Ø sloupc≈Ø | `Index(['cena', 'produkt'])` |
| `df.dtypes` | Datov√© typy sloupc≈Ø | `cena: int64, produkt: object` |

### üìå Jak si zapamatovat?
‚úî **`[]`** = indexov√°n√≠ (v√Ωbƒõr z kolekc√≠ jako list, dict, DataFrame)  
‚úî **`()`** = vol√°n√≠ funkc√≠ a metod  
‚úî **bez `()`** = atributy, kter√© vrac√≠ informace (ale nic nepoƒç√≠taj√≠)

### Hranat√© z√°vorky

Hranat√© z√°vorky vyu≈æijeme p≈ôi pr√°ci se seznamy.

```python
# Vytvo≈ôen√≠ nov√©ho seznamu
znamky = [2, 3, 2, 4, 2]
# P≈ôeƒçten√≠ hodnoty ze seznamu
posledni_znamka = znamky[-1]
# Ulo≈æen√≠ hodnoty na konkr√©tn√≠ pozici v seznamu
# Zde nap≈ô√≠klad zlep≈°√≠me posledn√≠ zn√°mku o jeden stupe≈à, proto≈æe student odevzdal v√Ωbornou p√≠semnou pr√°ci
znamky[-1] = posledni_znamka - 1
print(znamky)
```

Pomoc√≠ [slicing](https://kodim.cz/kurzy/uvod-do-progr-2/uvod-do-programovani-2/slicing-metody-moduly/slicing) m≈Ø≈æeme z√≠skat v√≠ce hodnot najednou - nap≈ô√≠klad t≈ôi hodnoty ze zaƒç√°tku. 

Slicing m√° smysl jen ve chv√≠li, kdy chci pouze ƒç√°st hodnot ze seznamu. Pokud chci cel√Ω seznam, nap√≠≈°u jen n√°zev promƒõnn√© a hranat√© z√°vorky nepou≈æ√≠v√°m.

```python
znamky = [2, 3, 2, 4, 2]
pocatecni_tri_znamky = znamky[:3]
print(pocatecni_tri_znamky)
```

```python
# Urƒçen√≠ poƒçtu zn√°mek pomoc√≠ funkce len
pocet_znamek = len(znamky)
# Tento ≈ô√°dek by zjistil to sam√© slicing je zbyteƒçn√Ω, pokud chci pracovat s cel√Ωm seznamem
# pocet_znamek = len(znamky[0:])
print(pocet_znamek)
```

Z√≠sk√°vat hodnotu pomoc√≠ hranat√Ωch z√°vorek m≈Ø≈æeme i z ≈ôetƒõzc≈Ø

```python
jmeno = "Jirka"
prijmeni = "Pe≈°√≠k"
inicialy = jmeno[0] + prijmeni[0]
```
